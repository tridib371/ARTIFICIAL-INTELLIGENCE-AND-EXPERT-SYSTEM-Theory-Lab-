# N Queen Attack Detection

```python
def drawBoard(P):
  N = len(P)
  for i in range(1,N+1):
    print("|", end="")
    for j in range(1,N+1):
      if (i,j) in P:
        ind = str(1+ P.index((i,j)))
        print(" "+ind+" |", end= "")
      else:
        print(" * |", end= "")
    print()
```

```python
import math
import random
def generate_positions(N):
    CL = list(range(0,N*N));
    P = []
    for i in range(N):
        p = random.choice(CL)
        CL.remove(p)
        r = math.ceil((p+1)/N)
        c = 1+p%N
        P.append((r,c))
        print('Q{}: {},{}'.format(i+1,r,c))
    return P

generate_positions(5)
```

```python
N = 5
positions = generate_positions(N)
drawBoard(positions)
```

```python
# row attack
def rowAttacK(P):
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if P[i][0] == P[j][0]:
        print('row attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count

rowAttacK(positions)
```

```python
#column attack
def columnAttacK(P):
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if P[i][1] == P[j][1]:
        print('Column attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count

columnAttacK(positions)
```

```python
# diagonal atack
def diagonalAttack(P, N):
    count = 0
    for i in range(N):
        for j in range(i + 1, N):
            if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):
                print('Diagonal attack:', end=" ")
                print(P[i], end=", ")
                print(P[j])
                count = count + 1
    return count
diagonalAttack(positions, 4)
```

```python
# implement a method to print a state is valid or not
# input row and col (two lists)
def isValid(P):

  row_attacks = rowAttacK(P)
  col_attacks = columnAttacK(P)
  diag_attacks = diagonalAttack(P, len(P))

  if row_attacks == 0 and col_attacks == 0 and diag_attacks == 0:
      return True
  else:
      return False
```

```python
# call isValid untill you get valid solution
while True:
    positions = generate_positions(4)
    print("Queens at:", positions)
    if isValid(positions):
        print("Valid Solution")
        break
    else:
        print("Invalid Solution")

```
