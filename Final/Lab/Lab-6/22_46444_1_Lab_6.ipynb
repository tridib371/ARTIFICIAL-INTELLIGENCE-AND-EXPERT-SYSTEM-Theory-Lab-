```python
def drawBoard(P):
  n = len(P)
  for i in range(1,n+1):
    print("|", end="")
    for j in range(1,n+1):
      if (i,j) in P:
        ind = str(1+ P.index((i,j)))
        print(" "+ind+" |", end= "")
      else:
        print(" * |", end= "")
    print()
```

```python
# diagonal atack
def diagonalAttack(P):
    n = len(P)
    count = 0
    for i in range(n):
        for j in range(i+1, n):
            if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):
                print('diagonal attack:', end=" ")
                print(i+1, P[i], end=", ")
                print(j+1, P[j])
                count = count + 1
    return count

diagonalAttack(pos)
```

```python
def isValid(P):
    # Check for row attack
    for i in range(len(P)):
        for j in range(i + 1, len(P)):
            if P[i][0] == P[j][0]:  # Same row
                print("Row attack detected between:", P[i], "and", P[j])
                return False

    # Check for column attack
    for i in range(len(P)):
        for j in range(i + 1, len(P)):
            if P[i][1] == P[j][1]:  # Same column
                print("Column attack detected between:", P[i], "and", P[j])
                return False

    # Check for diagonal attack
    for i in range(len(P)):
        for j in range(i + 1, len(P)):
            if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):  # Same diagonal
                print("Diagonal attack detected between:", P[i], "and", P[j])
                return False

    # If no attacks found, the state is valid
    return True
```

```python
# call isValid untill you get valid solution
import random

# Function to generate random positions for N Queens
def generate_positions(n):
    numbers = list(range(n**2))
    random.shuffle(numbers)
    P = []
    for i in range(n):
        row = numbers[i] // n + 1
        col = numbers[i] % n + 1
        P.append((row, col))
    return P

# Function to check if the position is valid
def isValid(P):
    # Check for row attack
    for i in range(len(P)):
        for j in range(i + 1, len(P)):
            if P[i][0] == P[j][0]:  # Same row
                print("Row attack detected between:", P[i], "and", P[j])
                return False

    # Check for column attack
    for i in range(len(P)):
        for j in range(i + 1, len(P)):
            if P[i][1] == P[j][1]:  # Same column
                print("Column attack detected between:", P[i], "and", P[j])
                return False

    # Check for diagonal attack
    for i in range(len(P)):
        for j in range(i + 1, len(P)):
            if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):  # Same diagonal
                print("Diagonal attack detected between:", P[i], "and", P[j])
                return False

    # If no attacks found, the state is valid
    return True

# Function to find a valid solution for N Queens
def findValidSolution(n):
    valid = False
    while not valid:
        pos = generate_positions(n)
        print(f"Generated positions: {pos}")
        valid = isValid(pos)  # Check if the generated solution is valid
    print("Valid solution found:", pos)
    return pos

# Example: Finding a valid solution for a 4x4 board
findValidSolution(4)
```
